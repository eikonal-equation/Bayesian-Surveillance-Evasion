```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
actual_cost = np.loadtxt('../output/Surveillance_Avoid/actual_cost.txt')

n = len(actual_cost)
plt.figure(figsize = (10, 10));
plt.plot(actual_cost);
plt.show();
```

```python
capture = np.loadtxt('../output/Surveillance_Avoid/capture_times.txt')

capture_rate = [capture[i] / (i + 1) for i in range(len(capture))]
n = len(actual_cost)
plt.figure(figsize = (10, 10));
plt.plot(capture_rate, linewidth = 3)
plt.xlim([0, len(capture)])
plt.ylim([0, 1])
plt.show();
```
